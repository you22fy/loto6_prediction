```python
import pandas as pd
```

```python
df = pd.read_csv('./loto6.csv', encoding="shift-jis")
df = df[['第1数字', '第2数字', '第3数字', '第4数字', '第5数字', '第6数字', 'BONUS数字']]
df
```

```python
counter = {k:0 for k in range(1,44)}
for i in range(len(df)):
    for item in list(df.iloc[i]):
        counter[item] += 1

for k, v in counter.items():
    print(f'{k} : {v}')
```

```python
import matplotlib.pyplot as plt
x = counter.keys()
y = counter.values()

plt.bar(x, y, width=0.75)
plt.show()
```

```python
sorted_counter = dict(sorted(counter.items(), key=lambda x: x[1], reverse=True))
for k, v in sorted_counter.items():
    print(f'{k} : {v}')
```
