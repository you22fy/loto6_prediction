```python
import pandas as pd
df = pd.read_csv('../loto6.csv', encoding="shift-jis")
df = df[['第1数字', '第2数字', '第3数字', '第4数字', '第5数字', '第6数字', 'BONUS数字']]
df
```

```python
frequency_map = {k:[0 for _ in range(len(df))] for k in range(1, 44)}
for i in range(1,44):
    for j in range(len(df)):
        if i in list(df.iloc[j]):
            frequency_map[i][j] = 1
for k, v in frequency_map.items():
    print(f'{k}:{v}')
```

```python
t = [i+1 for i in range(len(df))]
```

```python
import matplotlib.pyplot as plt
for i in range(1,44):
    plt.bar(t, frequency_map[i])
    file_name = f'No{i}'
    plt.savefig(f'./出現度別グラフ/{file_name}')
    plt.clf()
```
